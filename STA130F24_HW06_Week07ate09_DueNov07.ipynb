Linear regression model is used to describe the relationship between independent variable X and dependent variable Y by the equation $Y = \beta_0 + \beta_1 X + \epsilon $, where $\beta_0$ is the intercept, $\beta_1$ is the slope, and $\epsilon$ is a normally distributed error term with mean 0 and variance  $\sigma^2$. It implis that for any X, Y are normally distributed around the regression line, combining a linear trend with random variability.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define parameters for the theoretical model
beta_0 = 2  # Intercept
beta_1 = 3  # Slope
sigma = 1.5  # Standard deviation of the error term

# Generate values for the independent variable X
np.random.seed(0)  # For reproducibility
X = np.linspace(0, 10, 100)  # 100 evenly spaced points from 0 to 10

# Generate the error term epsilon from a normal distribution
epsilon = norm.rvs(loc=0, scale=sigma, size=X.size)

# Calculate Y based on the theoretical regression model
Y = beta_0 + beta_1 * X + epsilon

# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color="blue", label="Simulated Data")
plt.plot(X, beta_0 + beta_1 * X, color="red", label="Theoretical Regression Line (Y = β0 + β1X)")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Theoretical Simple Linear Regression Model")
plt.legend()
plt.show()

# Display the normal distribution for Y at a fixed X value
X_fixed = 5  # Arbitrary X value to show Y's distribution
Y_fixed = beta_0 + beta_1 * X_fixed  # Mean of Y at this X
Y_dist = norm(loc=Y_fixed, scale=sigma)

# Plot the distribution of Y at X = X_fixed
x_vals = np.linspace(Y_fixed - 3*sigma, Y_fixed + 3*sigma, 100)
plt.figure(figsize=(8, 5))
plt.plot(x_vals, Y_dist.pdf(x_vals), color="green", label=f"Distribution of Y at X = {X_fixed}")
plt.axvline(Y_fixed, color="red", linestyle="--", label="Mean of Y for this X")
plt.xlabel("Y")
plt.ylabel("Density")
plt.title(f"Normal Distribution of Y at X = {X_fixed}")
plt.legend()
plt.show()
````

````python
import statsmodels.formula.api as smf  # Library for statistical modeling in Python
import plotly.express as px  # this is a ploting library
import numpy as np
import pandas as pd

# Simulate data for X and Y
np.random.seed(0)
beta_0 = 2   # Intercept
beta_1 = 3   # Slope
sigma = 1.5  # Standard deviation of the error term

# Generate independent variable X and error term
X = np.linspace(0, 10, 100)
epsilon = np.random.normal(0, sigma, size=X.size)
Y = beta_0 + beta_1 * X + epsilon

# Combine into a DataFrame with specified column names
df = pd.DataFrame({'x': X, 'Y': Y})

# what are the following two steps doing?
model_data_specification = smf.ols("Y~x", data=df) # defines the model specification, where "Y ~ x" is the formula indicating Y as the dependent variable and x as the independent variable.
fitted_model = model_data_specification.fit() # fits the model to the data, estimating the coefficents for the intercept and slope.

# what do each of the following provide?
fitted_model.summary()  # full model summary, including coeficients, R-squred, p-values and diagonostics. 
fitted_model.summary().tables[1]  # coefficient table with estimates, errors, t-values and p-values
fitted_model.params  # estimate the intercept and slope
fitted_model.params.values  # array of intercept and slope values
fitted_model.rsquared  # R-squred, representing the proportion of variance in Y explained by X

# what two things does this add onto the figure?
df['Data'] = 'Data' # hack to add data to legend; Create a column called "Data" allow us to label the scatter plot points.
fig = px.scatter(df, x='x',  y='Y', color='Data', 
                 trendline='ols', title='Y vs. x')
# Adding `trendline='ols'` to `px.scatter` creates a fitted regression line based on ordinary least squares.

# This is essentially what above `trendline='ols'` does
fig.add_scatter(x=df['x'], y=fitted_model.fittedvalues,
                line=dict(color='blue'), name="trendline='ols'")

fig.show() # USE `fig.show(renderer="png")` FOR ALL GitHub and MarkUs SUBMISSIONS
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import plotly.express as px
from scipy.stats import norm

# Common Parameters
beta_0 = 2  # Intercept of the theoretical model
beta_1 = 3  # Slope of the theoretical model

# Function to create a dataset, fit a model, and plot both lines
def simulate_and_plot(X, sigma, scenario_name):
    # Generate error term epsilon with the specified standard deviation
    epsilon = np.random.normal(0, sigma, size=X.size)
    
    # Generate Y values based on the theoretical model
    Y = beta_0 + beta_1 * X + epsilon
    
    # Create DataFrame for statsmodels
    df = pd.DataFrame({'x': X, 'Y': Y})
    
    # Fit the model using statsmodels
    model_data_specification = smf.ols("Y ~ x", data=df)
    fitted_model = model_data_specification.fit()
    
    # Create theoretical line for comparison
    theoretical_Y = beta_0 + beta_1 * X
    
    # Plot using matplotlib for both lines
    plt.figure(figsize=(10, 6))
    plt.scatter(X, Y, color="blue", label="Simulated Data")
    plt.plot(X, theoretical_Y, color="red", label="Theoretical Line (Y = β0 + β1X)")
    plt.plot(X, fitted_model.fittedvalues, color="green", linestyle="--", label="Fitted Line (OLS)")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title(f"Scenario: {scenario_name}")
    plt.legend()
    plt.show()
    
    # Print model summary for additional insights
    print(f"Scenario: {scenario_name}")
    print(fitted_model.summary())
    print("\n")

# Scenario 1: Small Sample Size
np.random.seed(0)
X_small = np.linspace(0, 10, 10)  # Only 10 points
simulate_and_plot(X_small, sigma=1.5, scenario_name="Small Sample Size")

# Scenario 2: Larger Error Variance
X_standard = np.linspace(0, 10, 100)
simulate_and_plot(X_standard, sigma=5.0, scenario_name="Larger Error Variance")

# Scenario 3: Larger Sample Size
X_large = np.linspace(0, 10, 500)  # 500 points for a denser dataset
simulate_and_plot(X_large, sigma=1.5, scenario_name="Larger Sample Size")

# Scenario 4: Non-Uniformly Distributed X
X_random = np.sort(np.random.rand(100) * 10)  # Randomly distributed X values in [0, 10]
simulate_and_plot(X_random, sigma=1.5, scenario_name="Non-Uniformly Distributed X")
````

````python
import scipy.stats as stats
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import plotly.express as px

n,x_min,x_range,beta0,beta1,sigma = 20,5,5,2,3,5
x = stats.uniform(x_min, x_range).rvs(size=n)
errors = stats.norm(loc=0, scale=sigma).rvs(size=n)
Y = beta0 + beta1 * x + errors

df = pd.DataFrame({'x': x, 'Y': Y})
model_data_specification = smf.ols("Y~x", data=df) 
fitted_model = model_data_specification.fit() 

df['Data'] = 'Data' # hack to add data to legend 
fig = px.scatter(df, x='x',  y='Y', color='Data', 
                 trendline='ols', title='Y vs. x')

# This is what `trendline='ols'` is
fig.add_scatter(x=df['x'], y=fitted_model.fittedvalues,
                line=dict(color='blue'), name="trendline='ols'")

x_range = np.array([df['x'].min(), df['x'].max()])
y_line = beta0 + beta1 * x_range
fig.add_scatter(x=x_range, y=y_line, mode='lines',
                name=str(beta0)+' + '+str(beta1)+' * x', 
                line=dict(dash='dot', color='orange'))

# Add vertical lines for residuals
for i in range(len(df)):
    fig.add_scatter(x=[df['x'][i], df['x'][i]],
                    y=[fitted_model.fittedvalues[i], df['Y'][i]],
                    mode='lines',
                    line=dict(color='red', dash='dash'),
                    showlegend=False)
    
# Add horizontal line at y-bar
fig.add_scatter(x=x_range, y=[df['Y'].mean()]*2, mode='lines',
                line=dict(color='black', dash='dot'), name='y-bar')

fig.show() # USE `fig.show(renderer="png")` FOR ALL GitHub and MarkUs SUBMISSIONS
````

The OLS method selects the line by minimizing the sum of squared differences between the observed and predicted values, ensuring a line that best fits the data. The "squares" are used to avoid cancellation of positive and negative errors and to penalize larger errors more heavily, providing a more robust fit.

### 13. The "short" and "long" wait times are not "before and after" measurements so there are not natural pairs on which to base differences on which to do a "one sample" (paired differences) *hypothesis test*; but, we can do "two sample" hypothesis testing using a *permuation test*, or create a 95% *bootstrap confidence interval* for the difference in means of the two populations. 

### (A) Do a permuation test $\;H_0: \mu_{\text{short}}=\mu_{\text{long}} \; \text{ no difference in duration between short and long groups}$ by "shuffling" the labels
### (B) Create a 95% bootstrap confidence interval  by repeatedly bootstrapping within each group and applying *np.quantile(bootstrapped_mean_differences, [0.025, 0.975])* to the collection of differences between the sample means.    
### (a) Explain how the sampling approaches work for the two simulations.
### (b) Compare and contrast these two methods with the *indicator variable* based model approach used in Question 11, explaining how they're similar and different.<br>
    
<details class="details-example"><summary style="color:blue"><u>Further Guidance</u></summary>

> _You'll need to create `for` loops for repeated (shuffling simulation) **permutation** and (subgroup) **bootstrapping**, where_
>
> - _"shuffling" for **permutation testing** is done like this `old_faithful.assign(kind_shuffled=old_faithful['kind'].sample(n=len(old_faithful), replace=False).values)#.groupby('kind').size()`; then, the **mean difference statistic** is then calculated using `.groupby('kind_shuffled')['duration'].mean().iloc[::-1].diff().values[1]` (so the **observed statistic** is `old_faithful.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]`_
> 
>
> - _"two sample" **bootstrapping** is done like this `old_faithful.groupby('kind').apply(lambda x: x.sample(n=len(x), replace=True)).reset_index(drop=True)#.groupby('kind').size()`; then, the **bootstrapped mean difference statistic** is then calculated using `.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]` (like the **observed statistic** except this is applied to the **bootstrapped** resampling of `old_faithful`)_
> ---
>

````python
import numpy as np
import pandas as pd

# Define threshold for short and long waits
old_faithful['kind'] = np.where(old_faithful['waiting'] < 68, 'short', 'long')

# Calculate observed difference in means
observed_diff = old_faithful.groupby('kind')['duration'].mean().diff().iloc[-1]

# Permutation test
n_permutations = 10000
perm_diffs = []

for _ in range(n_permutations):
    shuffled_labels = old_faithful['kind'].sample(frac=1).reset_index(drop=True)
    permuted_data = old_faithful.assign(kind=shuffled_labels)
    mean_diff = permuted_data.groupby('kind')['duration'].mean().diff().iloc[-1]
    perm_diffs.append(mean_diff)

# Calculate p-value
perm_diffs = np.array(perm_diffs)
p_value = np.mean(np.abs(perm_diffs) >= np.abs(observed_diff))

print("Observed Difference in Means:", observed_diff)
print("P-value from Permutation Test:", p_value)

# Plot the permutation distribution
import matplotlib.pyplot as plt
import seaborn as sns

# Number of bootstrap samples
n_bootstraps = 10000
boot_diffs = []

# Bootstrap sampling for difference in means
for _ in range(n_bootstraps):
    short_sample = old_faithful[old_faithful['kind'] == 'short']['duration'].sample(frac=1, replace=True)
    long_sample = old_faithful[old_faithful['kind'] == 'long']['duration'].sample(frac=1, replace=True)
    mean_diff = long_sample.mean() - short_sample.mean()
    boot_diffs.append(mean_diff)

# Calculate 95% confidence interval
ci_lower, ci_upper = np.quantile(boot_diffs, [0.025, 0.975])

print("95% Bootstrap Confidence Interval for Difference in Means:", (ci_lower, ci_upper))

# Plot the bootstrap distribution
sns.histplot(boot_diffs, kde=True, color='salmon')
plt.axvline(ci_lower, color='blue', linestyle='--', label='Lower CI')
plt.axvline(ci_upper, color='blue', linestyle='--', label='Upper CI')
plt.xlabel('Mean Difference')
plt.title('Bootstrap Confidence Interval for Difference in Means')
plt.legend()
plt.show()
````

(a)Permutation Test: This approach uses label randomization to test the hypothesis of no difference. It's a non-parametric method that doesn't assume a particular distribution of the data.
Bootstrap Confidence Interval: This method relies on resampling within each group and using the distribution of resampled means to estimate the true difference. This approach also doesn’t rely on assumptions about the population distribution but instead uses the variability within the observed data to approximate the sampling distribution.

(b)Permutation and bootstrap methods are both non-parametric approaches but indicator variable model is a parametric method that assumes a linear relationship and normally distributed residuals. Indictor model can extend to more complex relationships by including additional predictors.

> - and the $\epsilon_i$ **errors** are **unbiased** relative to the **expected value** of **outcome** $E[Y_i|x_i]=\beta_0 + \beta_1x_i$ (which is equivalently stated by saying that the mean of the **error distribution** is $0$, or again equivalently, that the **expected value** of the **errors** $E[\epsilon_i] = 0$)
