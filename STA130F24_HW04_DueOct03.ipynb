````python
import numpy as np

# function to return the confidence interval for population median based on three parameter
def bootstrap_ci(data, num_samples=1000, confidence_level=0.95):
    
    # Datasize
    n = len(data)

    # Create a 1D array of length num_samples with unitialized values.
    bootstrapped_medians = np.empty(num_samples)

    # Generate 'num_samples' bootstrapping sample
    for i in range(num_samples):
        # Generate bootstrapping sample from the data
        # the number of sample draw from data is the datasize
        # sample can be repeatedly picked
        bootstrap_sample = np.random.choice(data, size=n, replace=True)
        
        # Calculate the median of the bootstrap sample
        # Append the median of this bootstrap sample to the bootstrapped_means array
        bootstrapped_medians[i] = np.median(bootstrap_sample)

    # Calculate the lower and upper percentiles for the confidence interval
    lower_percentile = (1 - confidence_level) / 2 * 100
    upper_percentile = (1 + confidence_level) / 2 * 100

    # Calculate the confidence interval
    # np.percentile method first sorts array in ascending order
    # then calculate the index corresponding to the desired percentile
    ci_lower = np.percentile(bootstrapped_medians, lower_percentile)
    ci_upper = np.percentile(bootstrapped_medians, upper_percentile)

    return ci_lower, ci_upper

# an example data
if __name__ == "__main__":
    # Sample data
    data = np.array([5, 7, 8, 9, 10, 12, 14, 16, 18])
    
    # Calculate the 95% bootstrap confidence interval
    lower_ci, upper_ci = bootstrap_ci(data)

    print(f"95% Bootstrap Confidence Interval: ({lower_ci:.2f}, {upper_ci:.2f})")
````

### Interaction Summary:

1. **Difference between Population Parameter and Sample Mean**:
   - **Population Parameter**: Describes a characteristic of the entire population (e.g., population mean \(\mu\)), a fixed value but often unknown.
   - **Sample Mean**: An estimate of the population mean based on a subset of the population (the sample). It varies from sample to sample.

2. **Population Parameter and Sample Mean in Confidence Intervals**:
   - **Sample Mean** is the point estimate used to construct the confidence interval.
   - The **Population Parameter** is the unknown value we aim to estimate with the interval. The confidence interval provides a range of plausible values for the population mean, centered around the sample mean.

3. **Process of Bootstrapping**:
   - Bootstrapping involves repeatedly resampling from an original dataset (with replacement) to create new samples. For each resampled dataset, a statistic (e.g., mean) is calculated. The distribution of these statistics is used to estimate the variability of the original statistic and construct confidence intervals or perform hypothesis testing.

4. **Purpose of Bootstrapping**:
   - Bootstrapping estimates the **sampling distribution** of a statistic without assuming a specific population distribution. It helps estimate confidence intervals, standard errors, and bias, especially when the sample size is small or parametric assumptions (like normality) are not valid.

5. **Definition of Non-Parametric**:
   - Non-parametric methods do not assume a specific distribution for the data. They are more flexible and are often used with small sample sizes, skewed data, ordinal data, or when the distribution is unknown. Non-parametric tests focus on ranks or signs rather than actual data values.

### 3. Quantitative Analysis: 
Using bootstrapping process to analyze sample and estimate the population parameter, then create 95% confidence interval for mean of the population.

````python
# Bootstrapping function
def bootstrap(data, n_bootstrap=1000, random_seed=42):
    np.random.seed(random_seed) #  ensuring that random number generation is reproducible in the code
    boot_means = [] # create an empty list to store the mean of each resampled dataset
    
    # Bootstrapping process: sample with replacement n_bootstrap times
    for _ in range(n_bootstrap):
        sample = np.random.choice(data, size=len(data), replace=True) #draw a sample which have the same size as original data from data with replacement
        boot_means.append(np.mean(sample)) # store the mean of sample
    
    return np.array(boot_means) # convert boot_means list into numpy array

# Run bootstrapping on the 'HealthImprovement' data
boot_means = bootstrap(df['HealthImprovement'], n_bootstrap=10000)

# Calculate the 95% confidence interval
confidence_interval = np.quantile(boot_means, [0.025, 0.975])

# Output the confidence interval
print(f"95% Confidence Interval for the mean health improvement: {confidence_interval[0]:.2f} to {confidence_interval[1]:.2f}")

# Plot the bootstrap distribution of the mean
plt.hist(boot_means, bins=30, edgecolor='k', alpha=0.7)
plt.axvline(confidence_interval[0], color='red', linestyle='dashed', linewidth=2)
plt.axvline(confidence_interval[1], color='red', linestyle='dashed', linewidth=2)
plt.title('Bootstrap Distribution of the Mean Health Improvement')
plt.xlabel('Mean Health Improvement')
plt.ylabel('Frequency')
plt.show()
````

As we can see on the data visualization (number line), the mean falls to the right of zero (greater than zero), and the confidence interval is 0.9 to 5.6, excluding 0.
